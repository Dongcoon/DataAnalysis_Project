## 지도 데이터

```python
map_data = 'map/SIG/sig.shp'
map = gpd.read_file(map_data,encoding='euc-kr')
map.tail()
```

```python
mdf = map.reset_index().drop(columns=['index'])
mdf['location'] = mdf.SIG_KOR_NM
```

## 변수 데이터

```python
data = pd.read_csv('data/unmatch_data.csv',encoding='cp949')
data.head()
```

```python
wdf = data.drop(columns=['Unnamed: 0'])
wdf.tail()
```

## 지도데이터, 변수데이터 결합

```python
test = pd.merge(mdf,wdf, how='outer',on='location')
```

```python
cond = (test.geometry != None)

convert = test[cond]
```

```python
sig_autumn = gpd.GeoDataFrame(convert,geometry=convert.geometry)
sig_autumn.set_crs = {'init':'epsg:5179'}
```

## shp 파일로 저장

```python
sig_autumn.to_file('map/match_data/sig_autumn.shp', driver='ESRI Shapefile', encoding='cp949')
```
