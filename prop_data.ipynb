## 병합데이터 가공

```python
start = []
climax = []

for i in range(len(total4.start)):
    start.append(total4.start[i].timetuple().tm_yday)
    climax.append(total4.climax[i].timetuple().tm_yday)
print(len(start),len(total4.start))

total4['start'] = start
total4['climax'] = climax
total4['month'] = total4.month.apply(lambda x:int(x))
```

```python
total5 = pd.DataFrame({'location':total4.location,'year':total4.year,'month':total4.month,'lon':total4.lon,'lat':total4.lat,\
             'height':total4.height,'avg_tem':total4.avg_tem,'precip':total4.precip,'wind':total4.wind,'min_tem':total4.min_tem,\
             'hPa':total4.hPa,'sun':total4.sun,'g_tem':total4.g_tem,'start':total4.start,'climax':total4.climax,})
```

```python
def outlier_iqr(data, column): 
    # lower(하한), upper(상한) 글로벌 변수 선언하기     
    global lower, upper    
    
    # 4분위 수 기준 지정하기     
    q1, q3 = np.quantile(data[column], 0.25), np.quantile(data[column], 0.75)          
    
    # IQR 계산하기     
    iqr = q3 - q1    
    
    # outlier cutoff(기준점) 계산하기     
    cut_off = iqr * 1.5          
    
    # lower와 upper 구분값 구하기     
    lower, upper = q1 - cut_off, q3 + cut_off     
    
    # print('IQR은',iqr, '이다.')     
    print('lower 기준값은', lower, '이다.')     
    print('upper 기준값은', upper, '이다.')    
    
    # 1사 분위와 4사 분위에 속해있는 데이터 각각 저장하기     
    data1 = data[data[column] > upper]     
    data2 = data[data[column] < lower]    
    
    # 이상값 총 개수 구하기
    return print('총 이상값 개수는', data1.shape[0] + data2.shape[0], '이다.')
```

```python
total5.columns
```

```python
# 이상치가 포함된 행을 제외하고 나머지 행 추출
for i in range(4,14+1):
    outlier_iqr(total5,total5.columns[i])
    where = (total5[f'{total5.columns[i]}'] >= lower) & (total5[f'{total5.columns[i]}'] <= upper)
    total6 = total5[where].reset_index().drop(columns=['index'])
```
