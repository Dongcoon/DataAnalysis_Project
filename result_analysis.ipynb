```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time

from statsmodels.formula.api import ols
```

```python
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
linear = pd.read_csv('regression/linear_result.csv',encoding='cp949')
ridge = pd.read_csv('regression/Ridge_result.csv',encoding='cp949')
lasso = pd.read_csv('regression/Lasso_result.csv',encoding='cp949')
```

```python
# lo = ['','','','','','', '' , '','','']
# #  
# for i in range(len(lo)):

area = ''

plt.title("ML ") 
plt.plot(linear.index,linear[area],label=f'linear = {round(linear[area].mean(),3)}')
plt.plot(lasso.index,lasso[area],label=f'lasso = {round(lasso[area].mean(),3)}')
plt.plot(ridge.index,ridge[area],label=f'ridge = {round(ridge[area].mean(),3)}')
plt.ylabel('error')
plt.xlabel('location')
plt.legend()
plt.grid()
# plt.ylim(280,325)
plt.show()
```
